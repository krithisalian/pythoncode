```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("employment_to_population_1979_to_2022.csv")
df
```

```python
x=df.drop(['total_population'],axis=1)
x
```

```python
y=df['total_population']
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
print("xtrain shape:",xtrain.shape)
print("xtest shape:",xtest.shape)
print("ytrain shape:",ytrain.shape)
print("ytest shape:",ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(xtrain,ytrain)
y_pred=regressor.predict(xtest)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
train_score=regressor.score(xtrain,ytrain)
test_score=regressor.score(xtest,ytest)
```

```python
print('Train Score (R-Squared): ',train_score)
print('Test Score (R-Squared)',test_score)
```
