```python
x=df.drop('Species',axis=1)
x
```

```python
y=df['Species']
y
```

```python
y_encode=pd.get_dummies(y)
y_encode
```

```python
x_encode=pd.get_dummies(x)
x_encode
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_encode,y_encode,test_size=0.25,random_state=100)
print("x train shape:",x_train.shape)
print("x test shape:",x_test.shape)
print("y train shape:",y_train.shape)
print("y test shape:",y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
train_prediction=model.predict(x_train)
train_prediction
```

```python
test_prediction=model.predict(x_test)
test_prediction
```

```python
train_score = model.score(x_train,y_train)
test_score = model.score(x_test,y_test)
print("R-squared value of train_score:",train_score)
print("R-squared value of test_score:",test_score)
```
